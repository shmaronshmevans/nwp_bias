```python
def format_df(df):
    new_df = pd.DataFrame()
    value_list = []
    for x, _ in df.iterrows():
        count = int(df.iloc[x]["COUNT"])
        value = df.iloc[x]["VALUE"]
        for n in np.arange(count):
            val = value
            value_list.append(val)
    new_df["VALUE"] = value_list
    return new_df
```

```python
import pandas as pd


def process_csv_files(path: str, station_list: list) -> None:
    """
    Reads multiple CSV files containing aspect data for different sites,
    calculates the percentage of each aspect for each site, and combines
    the data into a single DataFrame. The resulting DataFrame is filled
    with 0 for missing values, and then saved as a CSV file.

    Args:
        path (str): The path where the CSV files are located.
        station_list (list): A list of station names corresponding to each site.

    Returns:
        None
    """

    # Initialize an empty DataFrame to store the combined data.
    df_complete = pd.DataFrame()

    # Loop over the CSV files and combine them into a single DataFrame.
    for i in range(1, 127):
        # Read the CSV file and calculate the percentage of each aspect.
        df_x = pd.read_csv(f"{path}/NLCD_{i}.csv")
        df_x = df_x.assign(Percentage=lambda x: (x["COUNT"] / sum(df_x["COUNT"]) * 100))

        # Add the site number and pivot the DataFrame to have aspects as columns.
        df_x["site"] = i
        df_x = df_x.pivot(index="site", columns="VALUE", values="Percentage")

        # Concatenate the current DataFrame with the combined DataFrame.
        df_complete = pd.concat([df_complete, df_x])

    # Add the station names as a column in the DataFrame and fill missing values with 0.
    df_complete["station"] = station_list
    df_complete = df_complete.fillna(0)

    # Save the DataFrame as a CSV file.
    df_complete.to_csv("/home/aevans/nwp_bias/src/correlation/data/NLCD_nam.csv")
    return df_complete
```
