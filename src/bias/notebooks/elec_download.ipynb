```python
import pandas as pd
import numpy as np
import os
```

```python
path = "/home/aevans/nwp_bias/data/model_data/nyc_electricity/2022/"
files = os.listdir(path)
files = sorted(files)

master_df = pd.DataFrame()

for f in files:
    df = pd.read_csv(f"{path}/{f}")
    df = df[df["Name"] == "N.Y.C."]
    print(df)
    master_df = pd.concat([df, master_df])
```

```python
master_df = master_df.sort_values("Time Stamp")
```

```python
master_df.to_csv(
    "/home/aevans/nwp_bias/data/model_data/nyc_electricity/nyc_elec_load_july_aug_2022.csv"
)
```
